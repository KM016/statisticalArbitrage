```python
import pandas as pd
import numpy as np
```

```python
# load the dfs
one = pd.read_csv('data/wiki_daily_views_all.csv', parse_dates=['date'], index_col='date')
two = pd.read_csv('data/wiki_extras.csv', parse_dates=['date'], index_col='date')

# joining the dfs
merged = one.join(two, how='outer', lsuffix='_all', rsuffix='_manual')

# sort by date
merged = merged.sort_index()

# rename cols and save
merged = merged.rename(columns={'Steemit': 'STEEM', '0x_(decentralized_exchange_infrastructure)': 'ZRX', 'ICON_(blockchain_platform)': 'ICX'})
merged.to_csv('data/wiki_dv.csv')
```

```python
df = pd.read_csv('data/wiki_dv.csv')

df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date').sort_index()

# compute daily % change, clean up infs, drop rows where every coin is NaN
daily_change = df.pct_change()
daily_change = daily_change.replace([np.inf, -np.inf], np.nan).dropna(how='all')

daily_change.to_csv('data/wiki_daily_pct_change.csv')
```
