# Imports

```python
import pandas as pd
```

# Loading in data

```python
raw = pd.read_pickle('data/CoinGecko_top1k.pkl')
raw
```

# Cleaning

```python
price_cols = [col for col in raw.columns if 'price' in col]
coin_price = raw[price_cols]
coin_price.head()
```

```python
coin_price.columns = coin_price.columns.droplevel(1)
```

```python
coin_price.head()
```

```python
coin_price = coin_price.drop(columns=['TUSD', 'DAI', 'WBTC', 'WETH', 'USDC', 'BSV'])
```

filtering so we have only coins with ≥ 90% of data points

```python
# Calculate the total number of data points
tot_data_points = len(coin_price)
non_null_points = coin_price.notnull().sum()

thresh= 0.90 * tot_data_points

coin_price = coin_price.loc[:, non_null_points >= thresh]
coin_price.index = pd.to_datetime(coin_price.index, format='%Y-%m-%d')
coin_price.index = coin_price.index.strftime('%Y-%m-%d')
coin_price.head()
```

# Exporting

```python
coin_price.to_csv('data/coins.csv', index=True)
```
